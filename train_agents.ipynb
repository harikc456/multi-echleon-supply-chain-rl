```python
import torch
import torch.nn as nn
import torch.optim as optim
import or_gym
from collections import namedtuple, deque
import numpy as np
import random
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
from DQNAgent import *
from DDQNAgent import *
```

```python
env = or_gym.make('InvManagement-v1')
```

```python
BATCH_SIZE = 256
GAMMA = 0.7
EPS_START = 0.9
EPS_END = 0.05
EPS_DECAY = 200
TARGET_UPDATE = 10
lr = 2e-4
NUM_EPISODES = 250
```

```python
dqn_agent = DQNAgent(env, GAMMA, EPS_START ,lr, len(env.state), BATCH_SIZE)
dqn_agent.optimize_model(NUM_EPISODES)
```

```python
ddqn_agent = DDQNAgent(env, GAMMA, EPS_START ,lr, len(env.state), BATCH_SIZE)
ddqn_agent.optimize_model(NUM_EPISODES)
```

```python
plt.plot(dqn_agent.avg_losses)
plt.plot(ddqn_agent.avg_losses)
plt.legend(['DQN','DDQN'])
plt.savefig("comparison.png")
```
